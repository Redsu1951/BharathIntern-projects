```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("housing.csv")
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.info()
```

```python
from sklearn.model_selection import train_test_split
i=dataset.drop(['median_house_value'],axis=1)
j=dataset['median_house_value']
```

```python
i_train,i_test,j_train,j_test=train_test_split(i,j,test_size=0.2)
```

```python
train_data=i_train.join(j_train)
```

```python
sns.heatmap(train_data.corr())
```

```python
plt.figure(figsize=(11,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['total_rooms']+1)
```

```python
train_data.hist(figsize=(11,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data.join(pd.get_dummies(train_data.ocean_proximity))
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
plt.figure(figsize=(11,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
plt.figure(figsize=(11,8))
sns.scatterplot(x="latitude",y="longitude",data=train_data,hue="median_house_value",palette="Spectral")
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(11,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
i_train,j_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
```

```python
i_train_scale=scale.fit_transform(i_train)
reg=LinearRegression()
reg.fit(i_train_scale,j_train)
```

```python
testing=i_test.join(j_test)
testing['total_rooms']=np.log(testing['total_rooms']+1)
testing['total_bedrooms']=np.log(testing['total_bedrooms']+1)
testing['population']=np.log(testing['population']+1)
testing['households']=np.log(testing['total_rooms']+1)
testing=testing.join(pd.get_dummies(testing.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
testing['bedroom_ratio'] = testing['total_bedrooms'] / testing['total_rooms']
testing['household_rooms'] = testing['total_rooms'] / testing['households']

i_test,j_test=testing.drop(['median_house_value'],axis=1),testing['median_house_value']
```

```python
i_test_scale=scale.transform(i_test)
```

```python
reg.score(i_test_scale,j_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor();
forest.fit(i_train_scale,j_train)
```

```python
forest.score(i_test_scale,j_test)
```

```python
from sklearn.model_selection import GridSearchCV
para_grid={
    "n_estimators":[100,200,300],
    "min_samples_split":[2,4],
    "max_depth":[None,4,8]
}
grid_search =GridSearchCV(forest,para_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)
grid_search.fit(i_train_scale,j_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(i_test_scale,j_test)
```
